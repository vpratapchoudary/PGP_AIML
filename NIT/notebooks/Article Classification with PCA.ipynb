```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df = pd.read_json("res/NewsArticles.json")
df.head()
```

```python
df_vector = df['Vector']
df_vector.head()
```

```python
vectors = pd.DataFrame(list(df_vector))
vectors.head()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(vectors)
res1_sse = round(kmeans.inertia_, 2)
res1_sse
```

```python
df['LabelWithoutPCA'] = kmeans.labels_
counts = pd.DataFrame(df['LabelWithoutPCA'].value_counts().reset_index())
res3_hival_clus = counts.iloc[0,0]
res4_hival_count = counts.iloc[0,1]
print(res3_hival_clus)
print(res4_hival_count)
```

```python
for each,subset in df.groupby('LabelWithoutPCA'):
    print ("For Label :",each)
    print ( ' '.join(subset['Preprocessed-Article']).lower().split()[:50] )
res7 = 'stages'
```

```python
pca = PCA(n_components=100)
pca.fit(vectors.transpose())
pca_vectors = pd.DataFrame(pca.components_).transpose()
pca_vectors.head()
```

```python
kmeans_pca = KMeans(n_clusters=5, random_state=42)
kmeans_pca.fit(pca_vectors)
res2_sse = round(kmeans_pca.inertia_, 2)
res2_sse
```

```python
df['LabelWithPCA'] = kmeans_pca.labels_
counts_pca = pd.DataFrame(df['LabelWithPCA'].value_counts().reset_index())
res5_hival_clus = counts_pca.iloc[0,0]
res6_hival_count = counts_pca.iloc[0,1]
print(res5_hival_clus)
print(res6_hival_count)
```

```python
for each,subset in df.groupby('LabelWithPCA'):
    print ("For Label :",each)
    words = ( ' '.join(subset['Preprocessed-Article']).lower().split()[:50] )
    print(words)
    if each==2:
        res8 = words[49]
res8
#res8 = 'instead'
```
