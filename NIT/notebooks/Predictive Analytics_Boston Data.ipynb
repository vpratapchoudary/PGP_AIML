```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

np.random.seed(1)
X = np.random.normal(0, 1, 500)
X = X.reshape(500, 1)
sdev_err = np.sqrt(0.75)
err = np.random.normal(0.5, sdev_err, 500)
y = -1+0.5*X+err

#poly = PolynomialFeatures(degree = 2)
#X_poly = poly.fit_transform(X)
#X_poly = X_poly.reshape(500, 1)
#poly.fit(X_poly, y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

lin = LinearRegression()
lin.fit(X_train, y_train)

coeff = np.round(lin.coef_, decimals = 3)
incp = np.round(lin.intercept_, decimals = 3)

predicted = lin.predict(X_test)
mse = round(mean_squared_error(y_test, predicted), 2)

out_list = [[coeff], [incp], [mse]]

"""import csv
file = open("output/output.csv", mode = "w")
with file:
write = csv.writer(file)
write.writerows(out_list)
file.close()
"""
print(coeff)
print(incp)
print(mse)
print(predicted)
```

```python
#Boston data question
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from scipy import stats

#1
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df['MEDV'] = boston.target

#boston_df.head()

#boston_df.shape
n_rows = boston_df.shape[0]
n_columns = boston_df.shape[1]

#2
pop_mean = np.mean(boston_df['MEDV'])

#3
pop_std = np.std(boston_df['MEDV'])

#4
pop_se = pop_std/np.sqrt(len(boston_df["MEDV"]))

#5
#np.random.seed(1)
#boston_df['MEDV']

import random
random.seed(1)
samp_means = []
medv_list = boston_df['MEDV'].tolist()
samp_size = 30
for i in range(100):
    sample = random.sample(medv_list, samp_size)
    mean_sample = np.mean(sample)
    samp_means.append(mean_sample)
samp_dist = np.array(samp_means)
samp_dist_mean = np.mean(samp_dist)
samp_dist_std = np.std(samp_dist)    
samp_dist_se = pop_std/np.sqrt(samp_size)

#6
#import seaborn as sns
#sns.distplot(samp_means)

ci_lower = samp_dist_mean - (1.96*samp_dist_se)
ci_upper = samp_dist_mean + (1.96*samp_dist_se)

outlist1 = [[n_rows],[n_columns],[round(pop_mean,2)]]
outlist2 = [[round(pop_std,2)],[round(pop_se,2)]]
outlist3 = [[round(samp_dist_mean,1)],[round(samp_dist_std,1)],[round(samp_dist_se,1)],[round(ci_lower,2)],[round(ci_upper,2)]]

outlist = [outlist1, outlist2, outlist3]

import csv
for i in range(3):
    file = open("output/output" + str(i+1) + ".csv", mode = 'w', newline = '')
    with file:
        write = csv.writer(file)
        write.writerows(outlist[i])
    file.close()
```

```python
#wages question
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score

wages_df = pd.read_csv("res/wages.csv")

X = wages_df[['age']]
y = wages_df['wage']

degree_max = 4

cv_mean_scores = []

for i in range(1, degree_max+1):
    poly = PolynomialFeatures(i)
    X_poly = poly.fit_transform(X)
    slm_poly = LinearRegression()
    slm_poly.fit(X_poly, y)
    predict_poly = slm_poly.predict(X_poly)
    
    scores = cross_val_score(slm_poly, X_poly, y, cv = 5)
    cv_mean_scores.append([round(np.mean(scores),4)])
    
    #cv_mean_scores.append(cross_val_score(slm_poly, X_poly, y, cv = 5, scoring = 'neg_mean_squared_error'))

#outlist = cv_mean_scores

import csv
file = open("output/output.csv", mode = "w", newline = '')
with file:
    write = csv.writer(file)
    write.writerows(cv_mean_scores)
file.close()
    

```
